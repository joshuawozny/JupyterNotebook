<h1>Means</h1>

<h2>Population Mean</h2>  

$ \mu = \sum \frac{X_{i}}{N} $  
> $ \mu $ = population mean   
$ \sum X_{i} $ = sum of all scores present in the population: X1 X2 X3 and so on  
N = Total number of cases in a population  

<h2>Sample Mean</h2>  

$ \bar{x} = \sum \frac{x_{i}}{n} $  
> $ \bar{x} $ = sample mean   
$ \sum x_{i} $ = sum of all scores present in the sample: x1 x2 x3 and so on  
N = Total number of cases in a population  

```python
print("\nSample mean of body mass all species", round(penguins["Body Mass (g)"].mean(),2))
print("\nSample mean of body mass by", penguins["Body Mass (g)"].groupby(penguins["Species"]).mean().round(2))
```

<h1>Variances and Standard Deviations</h1>

>Represents how spread out the data is from the mean. Standard deviation is most often presented, particularly whith a normally distributed data set.

<h2>Population Variance</h2>

$ \sigma^2 = \frac{1}{N} \sum\limits _{i=1}^{N}(x_{i} - \mu)^2 $
> $ \sigma^2 $ = population variance   
N = size of the population consisting of $ x_{1}, x_{2}...x_{N} $  
$ \mu $ = population mean

<h2>Population Standard Deviation</h2>

$ \sigma = \sqrt{\frac{1}{N} \sum\limits _{i=1}^{N}(x_{i} - \mu)^2} $
> $ \sigma $ = population variance   
N = size of the population consisting of $ x_{1}, x_{2}...x_{N} $  
$ \mu $ = population mean

<h2>Sample Variance</h2>

$ s^2 = \frac{1}{n-1} \sum\limits _{i=1}^{n}(x_{i} - \bar{x})^2 $
> $ s^2 $ = sample variance   
n = size of the sample consisting of $ x_{1}, x_{2}...x_{n} $  
$ \bar{x} $ = sample mean

```python
print("\nSample variance of body mass all species", round(penguins["Body Mass (g)"].var(),2))
print("\nSample variance of body mass by", penguins["Body Mass (g)"].groupby(penguins["Species"]).var().round(3))
```

<h2>Sample Standard Deviation</h2>

$ s = \sqrt{\frac{1}{n-1} \sum\limits _{i=1}^{n}(x_{i} - \bar{x})^2} $
> $ s $ = sample standard deviation   
n = size of the sample consisting of $ x_{1}, x_{2}...x_{n} $  
$ \bar{x} $ = sample mean

```python
print("\nSample standard deviation of body mass all species", round(penguins["Body Mass (g)"].std(),2))
print("\nSample standard deviation of body mass by", penguins["Body Mass (g)"].groupby(penguins["Species"]).std().round(3))
```

<h2>Confidence Level</h2>

>The probability, with which the estimation of the location of a statistical parameter (e.g. an arithmetic mean) in a sample survey is also true for the population.

<h2>Confidence Interval (CI)</h2>

>The range of estimates for an unknown parameter. A confidence interval is computed at a designated confidence level. For instance, a confidence interval at the 95% confidence level is the range of values that will occur in the population 95% of the time.

```python
def print_conf_interval(data,conf_lvl=.95, type=1,round_dig=2):
    #type = 0 population, 1 sample
    if type == 0:
        descr = "population"    
    else:
        descr = "sample"   
    
    length = data.count() - type
    mean = data.mean()
    std_dev = data.std()
    interval = st.t.interval(conf_lvl,length,mean,std_dev)
    print_interval = (round(interval[0],round_dig),round(interval[1],round_dig))
    
    print("Confidence interval for", descr," at",round(conf_lvl*100,1),"% confidence: ",print_interval)
```

```python
data = penguins["Body Mass (g)"]
print_conf_interval(data)
```

```python
data = penguins.where(penguins.Species=="Adelie Penguin (Pygoscelis adeliae)")["Body Mass (g)"]
print_conf_interval(data,.9,1,3)
```
